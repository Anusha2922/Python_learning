```python
df['standardized age'] = (df['age']-df['age'].mean())/df['age'].std()
df['normalized age'] = (df['age']-df['age'].min())/(df['age'].max()-df['age'].min())
df['standardized BMI'] = (df['BMI']-df['BMI'].mean())/df['BMI'].std()
df['normalized BMI'] = (df['BMI']-df['BMI'].min())/(df['BMI'].max()-df['BMI'].min())
```

```python
for col in ['age','BMI']:
  df[f'standardized {col}'] = (df[col]-df[col].mean())/df[col].std()
```

Train test Split

```python
#seperate the features and target variable
X = df.drop(['outcome','group','ID'], axis=1)
y = df['outcome']
```

```python
# Filter out rows with unexpected values in the 'outcome' column
df_filtered = df[df['outcome'].isin([0.0, 1.0])].copy()

# Separate the features and target variable
X = df_filtered.drop(['outcome','group','ID'], axis=1)
y = df_filtered['outcome']

#split the data into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

Logical regression

```python
#logistic regression model to predict the outcome
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
y_pred = logreg.predict(X_test)
```

```python
#evaluate the model
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
#print these scores
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 score:", f1_score(y_test, y_pred))
```
